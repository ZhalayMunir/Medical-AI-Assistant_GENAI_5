```python
def extract_data_from_json(file_path):
    with open(file_path, 'r') as f:
        data = json.load(f)

    extracted_data = []

    def traverse_json(d, path=""):
        if isinstance(d, dict):
            for key, value in d.items():
                new_path = f"{path} > {key}" if path else key
                if isinstance(value, dict):
                    traverse_json(value, new_path)
                else:
                    extracted_data.append((new_path, value))

    traverse_json(data)
    return extracted_data
```

```python
import os
import json
import pandas as pd

# Root folder containing all disease subfolders
root_folder = "Finished/Finished"

# Function to process each JSON file and extract data

def extract_medical_data(file_path, disease_name):
    with open(file_path, "r", encoding="utf-8") as f:
        json_data = json.load(f)

    # Initialize a list to store the records
    records = []

    # Extract input values
    input_values = {k: v.strip() for k, v in json_data.items() if k.startswith("input")}

    # Extract additional data from input fields
    extracted_data = {
        'symptoms': input_values.get('input1', 'N/A'),
        'patient_history': input_values.get('input2', 'N/A'),
        'family_medical_history': input_values.get('input4', 'N/A'),
        'physical_exam': input_values.get('input5', 'N/A'),
        'laboratory_results': input_values.get('input6', 'N/A'),
    }

    # Iterate through the diagnoses and their associated causes and symptoms
    for diagnosis, causes in json_data.items():
        if isinstance(causes, dict):
            diagnosis_name = diagnosis.split("$")[0]
            for cause_desc, symptoms in causes.items():
                cause_text = cause_desc.split("$")[0]
                if isinstance(symptoms, dict):
                    for symptom_desc, _ in symptoms.items():
                        symptom_text = symptom_desc.split("$")[0]
                        input_ref = symptom_desc.split("$")[-1]
                        test_results = input_values.get(input_ref, "N/A")

                        # Combine the extracted data into a single record
                        combined_record = {
                            "Disease": disease_name,
                            "Diagnosis": diagnosis_name,
                            "Cause": cause_text,
                            "Symptom": symptom_text,
                            "Test Results": test_results,
                            "File": os.path.basename(file_path)  # Store file name for reference

                        }

                        # Update the combined record with the extracted input data
                        combined_record.update(extracted_data)

                        # Append the combined record to the records list
                        records.append(combined_record)

    return records

# Traverse all subfolders and process JSON files
all_data = []



for dirpath, dirnames, filenames in os.walk(root_folder):
    for file_name in filenames:
        if file_name.endswith(".json"):
            file_path = os.path.join(dirpath, file_name)

            # Extract disease name from the relative path after root_folder
            relative_path = os.path.relpath(dirpath, root_folder)
            disease_name = relative_path.split(os.sep)[0]  # top-level folder = main disease

            all_data.extend(extract_medical_data(file_path, disease_name))


# Convert to DataFrame
df = pd.DataFrame(all_data)

# Save to CSV
output_path = "medical_data.csv"
df.to_csv(output_path, index=False)

print(f"Data extracted and saved to {output_path}")
print(df.head(10))  # Display first 10 rows
df["Disease"].unique()

```

```python
import os
import json
import csv

def extract_data_from_json(folder_path, output_csv):
    data_list = []

    for file_name in os.listdir(folder_path):
        if file_name.endswith(".json"):
            file_path = os.path.join(folder_path, file_name)
            with open(file_path, "r", encoding="utf-8") as file:
                json_data = json.load(file)

            disease_name = os.path.splitext(file_name)[0]  # Extract filename without extension

            for main_category, subcategories in json_data.get("diagnostic", {}).items():
                for subcategory, details in subcategories.items():
                    # Extract knowledge section
                    knowledge_data = json_data.get("knowledge", {}).get(main_category, {})

                    if isinstance(knowledge_data, dict):
                        risk_factors = knowledge_data.get("Risk Factors", "N/A")
                        symptoms = knowledge_data.get("Symptoms", "N/A")
                        signs = knowledge_data.get("Signs", "N/A")
                    else:
                        risk_factors, symptoms, signs = "N/A", "N/A", "N/A"

                    # Some subcategories in knowledge might be stored as a string
                    subcategory_info = json_data.get("knowledge", {}).get(subcategory, "N/A")

                    data_list.append([
                        disease_name, main_category, subcategory,
                        risk_factors, symptoms, signs, subcategory_info
                    ])

    # Write to CSV
    with open(output_csv, "w", newline="", encoding="utf-8") as csv_file:
        writer = csv.writer(csv_file)
        writer.writerow(["Disease", "Main Category", "Subcategory",
                         "Risk Factors", "Symptoms", "Signs", "Subcategory Information"])
        writer.writerows(data_list)

# Example Usage
folder_path = "Diagnosis_flowchart/Diagnosis_flowchart"  # Update with the correct folder path
output_csv = "disease_knowledge.csv"
extract_data_from_json(folder_path, output_csv)
print(f"Data extracted and saved to {output_csv}")
```

```python
import os
import json
import pandas as pd

# Path to the diagnosis_flochart folder
folder_path = "Diagnosis_flowchart/Diagnosis_flowchart"

# Function to extract information from each JSON file
def extract_diagnosis_info(file_path):
    with open(file_path, "r", encoding="utf-8") as f:
        json_data = json.load(f)

    records = []

    # Extract "diagnostic" hierarchy (Diagnosis Flowchart)
    diagnostic_tree = json_data.get("diagnostic", {})

    # Extract "knowledge" section (Risk Factors, Symptoms, Tests)
    knowledge_data = json_data.get("knowledge", {})

    for main_disease, sub_diseases in diagnostic_tree.items():
        for sub_disease, subtypes in sub_diseases.items():
            diagnosis_hierarchy = f"{main_disease} → {sub_disease}"

            # Extract relevant medical knowledge
            risk_factors = knowledge_data.get(main_disease, {}).get("Risk Factors", "N/A")
            symptoms = knowledge_data.get(main_disease, {}).get("Symptoms", "N/A")
            signs = knowledge_data.get(main_disease, {}).get("Signs", "N/A")
            diagnostic_tests = knowledge_data.get(sub_disease, "N/A")

            records.append({
                "Main Disease": main_disease,
                "Sub Disease": sub_disease,
                "Diagnosis Flowchart": diagnosis_hierarchy,
                "Risk Factors": risk_factors,
                "Symptoms": symptoms,
                "Signs": signs,
                "Diagnostic Tests": diagnostic_tests,
                "File": os.path.basename(file_path)
            })

    return records

# Traverse all JSON files and extract data
all_data = []

for file_name in os.listdir(folder_path):
    if file_name.endswith(".json"):  # Process only JSON files
        file_path = os.path.join(folder_path, file_name)
        all_data.extend(extract_diagnosis_info(file_path))

# Convert extracted data to a Pandas DataFrame
df = pd.DataFrame(all_data)

# Save the extracted data to a CSV file
output_path = "diagnosis_data.csv"
df.to_csv(output_path, index=False)

print(f"Data extracted and saved to {output_path}")
print(df.head(10))  # Display first 10 rows
df["Main Disease"].nunique()

```

```python
import pandas as pd
import re
import numpy as np

# Load the datasets
medical_df = pd.read_csv("medical_data.csv")
diagnosis_df = pd.read_csv("diagnosis_data.csv")

# Function to clean text
def clean_text(text):
    if pd.isna(text):
        return "Unknown"
    text = re.sub(r'\s+', ' ', str(text)).strip()  # Remove extra spaces
    text = text.lower()  # Convert to lowercase
    return text

# Apply text cleaning
for col in medical_df.columns:
    medical_df[col] = medical_df[col].apply(clean_text)

for col in diagnosis_df.columns:
    diagnosis_df[col] = diagnosis_df[col].apply(clean_text)

# Drop irrelevant columns (example: test results)
if "Test Results" in medical_df.columns:
    medical_df.drop(columns=["Test Results"], inplace=True)
# Remove the word "Suspected" from the Disease column in diagnosis_df
diagnosis_df["Main Disease"] = diagnosis_df["Main Disease"].str.replace(r'\bsuspected\b', '', regex=True).str.strip()


# Save the cleaned versions
medical_df.to_csv("cleaned_medical_data.csv", index=False)
diagnosis_df.to_csv("cleaned_diagnosis_data.csv", index=False)

print("✅ Data cleaned and saved!")
```

```python
from sentence_transformers import SentenceTransformer
import faiss

# Load model
embedder = SentenceTransformer('all-MiniLM-L6-v2')

# Get the text corpus
corpus = merged_df["combined_text"].tolist()

# Generate embeddings
corpus_embeddings = embedder.encode(corpus, convert_to_numpy=True, show_progress_bar=True)

# Normalize for cosine similarity
corpus_embeddings = corpus_embeddings / np.linalg.norm(corpus_embeddings, axis=1, keepdims=True)
```

```python
dimension = corpus_embeddings.shape[1]
index = faiss.IndexFlatIP(dimension)  # Using cosine similarity after normalization

# Add all vectors
index.add(corpus_embeddings)
```

```python
def faiss_search(query, top_n=5):
    # Embed the query
    query_embedding = embedder.encode([query], convert_to_numpy=True)
    query_embedding = query_embedding / np.linalg.norm(query_embedding, axis=1, keepdims=True)

    # Search in the FAISS index
    distances, indices = index.search(query_embedding, top_n)

    # Get results from DataFrame
    return merged_df.iloc[indices[0]][[
        "Diagnosis", "Symptom", "Cause", "symptoms", "family_medical_history", "laboratory_results"
    ]]
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

model_name = "mistralai/Mistral-7B-Instruct-v0.2"  # Or other LLaMA-based model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name, device_map="auto", torch_dtype=torch.float16)


```

```python
def generate_answer(query, retrieved_docs):
    context = ""
    for idx, row in retrieved_docs.iterrows():
        parts = []
        for col in ["Diagnosis", "Symptom", "Cause", "symptoms", "family_medical_history", "laboratory_results"]:
            value = str(row.get(col, '')).strip()
            if value:
                parts.append(f"{col}: {value}")
        context += " | ".join(parts) + "\n"

    prompt = (
        f"You are a knowledgeable medical assistant. Based on the patient's query and the medical context below, "
        f"generate a long, detailed, structured and medically accurate response:\n\n"
        f"Query: {query}\n\n"
        f"Context:\n{context}\n\n"
        f"Response:"
    )

    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    outputs = model.generate(
        **inputs,
        max_new_tokens=400,
        temperature=0.7,
        top_p=0.9,
        repetition_penalty=1.2,
        eos_token_id=tokenizer.eos_token_id
    )
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Only keep the part after the word "Response:"
    if "Response:" in response:
        response = response.split("Response:", 1)[-1]

    return response.strip()
```

```python
def evaluate_generation(generated_texts, reference_texts):
    # BLEU needs tokenized words
    references = [[ref.split()] for ref in reference_texts]
    predictions = [gen.split() for gen in generated_texts]

    # ROUGE uses full strings
    rouge_result = rouge.compute(predictions=generated_texts, references=reference_texts)
    bleu_result = bleu.compute(predictions=predictions, references=references)

    print("📊 Evaluation Results:")
    print("🔹 ROUGE-1: {:.4f}".format(rouge_result['rouge1']))
    print("🔹 ROUGE-2: {:.4f}".format(rouge_result['rouge2']))
    print("🔹 ROUGE-L: {:.4f}".format(rouge_result['rougeL']))
    print("🔹 BLEU Score: {:.4f}".format(bleu_result['bleu']))
```

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch

# Load pre-trained GPT-2 model and tokenizer once
model_metric = GPT2LMHeadModel.from_pretrained("gpt2")
gpt2_tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model_metric.eval()  # Set model to evaluation mode

def calculate_perplexity(generated_response: str) -> float:
    inputs = gpt2_tokenizer(generated_response, return_tensors="pt", truncation=True, max_length=1024)
    labels = inputs.input_ids

    with torch.no_grad():
        outputs = model_metric(**inputs, labels=labels)
        loss = outputs.loss
        perplexity = torch.exp(loss)

    return perplexity.item()

perplexity = calculate_perplexity(generated_response)
print(f"Perplexity: {perplexity}")
```
